```python
#Day 11

#Part 1
import copy

f = open('Day_11_data.txt').read().splitlines()

def create_seatmap(input):
    seatmap = []
    for row in input:
        row_x = [seat for seat in row]
        seatmap.append(row_x)
    return seatmap

seatmap = create_seatmap(f)

def find_adjacent_seats(seat_map):
    adjacent_seats_dict = dict()
    for col in range(0,len(seat_map[0])):
        for row in range(0, len(seat_map)):
            #top left corner
            if col == 0 and row == 0:
                adjacent_seats = [seat_map[row][col+1],
                                seat_map[row+1][col],
                                seat_map[row+1][col+1]]
            #left side
            elif col == 0 and row in range (1,len(seat_map)-1):
                adjacent_seats = [seat_map[row][col+1],
                                seat_map[row-1][col],
                                seat_map[row+1][col],
                                seat_map[row+1][col+1],
                                seat_map[row-1][col+1]]
            #right side
            elif col == len(seat_map[0])-1 and row in range (1,len(seat_map)-1):
                adjacent_seats = [seat_map[row][col-1],
                                seat_map[row-1][col],
                                seat_map[row+1][col],
                                seat_map[row+1][col-1],
                                seat_map[row-1][col-1]]
            #bottom left corner
            elif col == 0 and row == len(seat_map)-1:
                adjacent_seats = [seat_map[row][col+1],
                                seat_map[row-1][col],
                                seat_map[row-1][col+1]]
            #bottom right corner
            elif col == len(seat_map[0])-1 and row == len(seat_map)-1:
                adjacent_seats = [seat_map[row][col-1],
                                seat_map[row-1][col],
                                seat_map[row-1][col-1]]
            #top row
            elif col in range(1,len(seat_map[0])-1) and row ==0:
                adjacent_seats = [seat_map[row+1][col+1],
                                seat_map[row][col-1],
                                seat_map[row+1][col],
                                seat_map[row+1][col-1],
                                seat_map[row][col+1]]
            #bottom row
            elif col in range(1,len(seat_map[0])-1) and row ==len(seat_map)-1:
                adjacent_seats = [seat_map[row-1][col+1],
                                seat_map[row][col-1],
                                seat_map[row-1][col],
                                seat_map[row-1][col-1],
                                seat_map[row][col+1]]
            #top right corner
            elif col == len(seat_map[0])-1 and row == 0:
                adjacent_seats = [seat_map[row][col-1],
                                seat_map[row+1][col],
                                seat_map[row+1][col-1]]
            else:
                adjacent_seats = [seat_map[row][col-1],
                                seat_map[row][col+1],
                                seat_map[row-1][col],
                                seat_map[row+1][col],
                                seat_map[row+1][col-1],
                                seat_map[row-1][col-1],
                                seat_map[row+1][col+1],
                                seat_map[row-1][col+1]]
            adjacent_seats_dict[(row,col)] =  adjacent_seats   
    return adjacent_seats_dict


def implement_seatmap_rules(seat_map):
    adjacent_seats_dict = find_adjacent_seats(seat_map)
    for key in adjacent_seats_dict.keys():
        if seat_map[key[0]][key[1]] == 'L' and adjacent_seats_dict[key].count('#') == 0:
            seat_map[key[0]][key[1]] = '#'
        if seat_map[key[0]][key[1]] == '#' and adjacent_seats_dict[key].count('#') >= 4:
            seat_map[key[0]][key[1]] = 'L'
    return seat_map
#Problem, need to find all adjacent seats first, THEN do the edits. Otherwise it's updated.
            

def find_seating_stability(original_seat_map):
    prior_seatmap = copy.deepcopy(original_seat_map)
    current_seatmap = implement_seatmap_rules(original_seat_map)
    while prior_seatmap != current_seatmap:
        prior_seatmap = copy.deepcopy(original_seat_map)
        current_seatmap = implement_seatmap_rules(original_seat_map)
    count_occupied = 0
    for row in range(0, len(current_seatmap)):
        count_occupied += current_seatmap[row].count('#')
    return count_occupied


find_seating_stability(seatmap)
```

```python
#Part 2
data = open('day_11_data.txt').read().splitlines()

def create_seatmap(input):
    seatmap = []
    for row in input:
        row_y = [seat for seat in row]
        seatmap.append(row_y)
    return seatmap

seatmap = create_seatmap(data)


def is_seat_valid(seat_map, row, col):
    return row in range (0,len(seat_map)) and col in range (0,len(seat_map[0]))

def count_adjacents(seat_map, row, col):
    slopes = {-1, 1, 0}
    count_occupied = 0
    for slope_x in slopes:
        for slope_y in slopes:
            #print('Slope is '+ str((slope_y, slope_x)))
            if slope_x == 0 and slope_y == 0:
                continue
            else:
                del_x = slope_x
                del_y = slope_y
                while is_seat_valid(seat_map, row + del_y, col + del_x) is True:
                    #print(count_occupied)
                    #print('Seat checked is ' + str((row + del_y , col + del_x)))
                    if seat_map[row + del_y][col + del_x] == 'L':
                        break
                    if seat_map[row + del_y][col + del_x] == '#':
                        count_occupied += 1
                        break
                    if seat_map[row + del_y][col + del_x] == '.':
                        del_y = del_y + slope_y
                        del_x = del_x + slope_x
                    
    return count_occupied


def create_seat_dictionary(seat_map):
    count_of_occupied_dict = dict()
    for col in range(0,len(seat_map[0])):
        for row in range(0, len(seat_map)):
            count_of_occupied_dict[(row, col)] = 0
    return count_of_occupied_dict
             

def enact_round(seat_map, dict):
    has_changed = False
    for col in range(0,len(seat_map[0])):
        for row in range(0, len(seat_map)):
            dict[(row, col)] = count_adjacents(seat_map, row, col)
    for seat in dict:
        if seat_map[seat[0]][seat[1]] == '#' and dict[seat] >= 5: 
            seat_map[seat[0]][seat[1]] = 'L'
            has_changed = True
        elif seat_map[seat[0]][seat[1]] == 'L' and dict[seat] == 0: 
            seat_map[seat[0]][seat[1]] = '#'
            has_changed = True
    return has_changed



def find_stabilization_count_occupied(seat_map):
    seat_dictionary = create_seat_dictionary(seat_map)
    prior_seatmap = copy.deepcopy(seat_map)
    has_changed = enact_round(seat_map, seat_dictionary)
    while has_changed == True:
        prior_seatmap = copy.deepcopy(seat_map)
        has_changed = enact_round(seat_map, seat_dictionary)
    count_occupied = 0
    for row in range(0, len(seat_map)):
        count_occupied += seat_map[row].count('#')
    return count_occupied
    

#Execute

# print(seatmap[2])
find_stabilization_count_occupied(seatmap)


                

```
